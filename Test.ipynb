```python
import pandas as pd
import json
from bs4 import BeautifulSoup
# Load data into DataFrame
df = pd.read_csv("D:\GenAI\Sample_data.csv")

df['contact_info'] = df['contact_info'].apply(lambda x: json.loads(x) if isinstance(x, str) else [])
df['facilities'] = df['facilities'].apply(lambda x: json.loads(x) if isinstance(x, str) else {})


# Remove HTML tags from the details column
df['details'] = df['details'].apply(lambda x: BeautifulSoup(x, "html.parser").get_text())
# Create DataFrame for exploded contact_info and facilities
contact_info_df = pd.json_normalize(df['contact_info'].explode())
facilities_df = pd.json_normalize(df['facilities'])

# Merge exploded DataFrames back to the original DataFrame
df = df.drop(columns=['contact_info', 'facilities'])
df = df.join(contact_info_df).join(facilities_df)

# Check the updated DataFrame
df.head()
```

```python
df.to_csv("D:\GenAI\Sample_data_.csv",index=False)
```

```python
import sqlite3
import pandas as pd

# Function to insert CSV data into SQLite
def insert_csv_to_sqlite(csv_file, sqlite_db, table_name):
    # Connect to SQLite database (it will create the database if it doesn't exist)
    conn = sqlite3.connect(sqlite_db)
    cursor = conn.cursor()

    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # Insert the DataFrame into the SQLite database
    df.to_sql(table_name, conn, if_exists='replace', index=False)

    # Commit and close the connection
    conn.commit()
    conn.close()
    print(f"Data from {csv_file} inserted into {table_name} table in {sqlite_db} database successfully.")

# Example usage
csv_file = 'Example_data.csv'  # Replace with your CSV file path
sqlite_db = 'database.db'  # Replace with your SQLite database file path
table_name = 'sample_data'  # Replace with your desired table name

insert_csv_to_sqlite(csv_file, sqlite_db, table_name)
```

```python
import sqlite3

# Function to query data from SQLite and print it
def query_sqlite(sqlite_db, query):
    # Connect to SQLite database
    conn = sqlite3.connect(sqlite_db)
    cursor = conn.cursor()

    # Execute the query
    cursor.execute(query)

    # Fetch all rows from the executed query
    rows = cursor.fetchall()

    # Print the fetched rows
    for row in rows:
        print(row)

    # Close the connection
    conn.close()

# Example usage
sqlite_db = 'database.db'  # Replace with your SQLite database file path
query = "SELECT * FROM sample_data WHERE name='OKA HAUS Sukhumvit 36'"  # Replace with your query
# SELECT * FROM sample_data WHERE 1=1 AND name = 'OKA HAUS Sukhumvit 36
query_sqlite(sqlite_db, query)
```
